```python
#Checking the distribution of the Class according to the Amount of transaction associated
target_var="Class"
dependant_var="Amount"
operation_list={
    "Count":"count",
    "Min":"min",
    "Max":"max",
    "Sum":"sum",
    
}

def groupby_info_target_var(df,target_var,dependant_var,operations):
    groupby_tab={"Class":[0,1]}
    for operation_name, operation in operations.items():
        res=df.groupby([target_var])[dependant_var].agg(operation)
        groupby_tab[f"{operation_name}"]=[res[0],res[1]]
    groupby_tab=pd.DataFrame(groupby_tab)
    return groupby_tab

groupby_tab=groupby_info_target_var(df,target_var,dependant_var,operation_list)

groupby_tab
```

### We will use the following ensemble models from imblearn.ensemble : `BalancedRandomForestClassifier` , `EasyEnsembleClassifier`

```python
from imblearn.ensemble import EasyEnsembleClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.metrics import classification_report,roc_auc_score,roc_curve,precision_recall_curve,auc
from sklearn.model_selection import train_test_split
```

## a- Data preparation

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

## b- Model setting up and evaluation 

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve, classification_report
from imblearn.ensemble import EasyEnsembleClassifier, BalancedRandomForestClassifier
import matplotlib.pyplot as plt

models = {
    "EasyEnsembleClassifier": EasyEnsembleClassifier(),
    "BalancedRandomForestClassifier": BalancedRandomForestClassifier(),
}

evaluations = []

for model_name, model in models.items():
    evals = {}
    model.fit(xtrain, ytrain)
    
    ypred = model.predict(xtest)
    ypred_proba = model.predict_proba(xtest)[:, 1]  # Use probability of the positive class
    
    # ROC Curve and AUC
    fpr, tpr, _ = roc_curve(ytest, ypred_proba)
    roc_auc = auc(fpr, tpr)
    
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic {model_name}')
    plt.legend(loc="lower right")
    plt.show()
    
    # Precision-Recall Curve and AUC
    precision, recall, _ = precision_recall_curve(ytest, ypred_proba)
    pr_auc = auc(recall, precision)
    
    evals["Precision"], evals["Recall"] = precision, recall
    evals["Precision Recall AUC"] = pr_auc
    
    plt.figure()
    plt.plot(recall, precision, color='blue', lw=2, label='Precision-Recall curve (area = %0.2f)' % pr_auc)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall Curve {model_name}')
    plt.legend(loc="lower left")
    plt.show()
    
    evaluations.append((model_name, evals))

for model_name, evals in evaluations:
    print(f"Model: {model_name}")
    print(f"Precision-Recall AUC: {evals['Precision Recall AUC']:.2f}")
    print(classification_report(ytest, models[model_name].predict(xtest)))
```

## c- Interpreation

Since our dataset is highly imbalanced the more informatives metrics are Precisions-Recall AUC.

`Precision-Recall AUC`:  Both models (0.83-> EassyEnsembleClassifier , 0.82 -> BalancedRandomForestClassifier)  perform well in distinguishing between the classes,particularly in the minority class(positive class)
